```python
import pandas as pd
import roman
```

```python
df1 = pd.read_csv('raw_ricky_morty.csv')
```

```python
df1.sort_values('Name')
```

```python
df1['Name'] = df1['Name'] + ' ' + df1.groupby('Name').cumcount().apply(lambda x: roman.toRoman(x+1) if x > 0 else '')
```

```python
df1.to_csv('cleaned_rick_morty.csv', index= False)
```
